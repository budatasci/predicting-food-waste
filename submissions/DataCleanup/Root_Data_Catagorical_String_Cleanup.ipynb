```python
import pandas as pd
import re
df = pd.read_csv("BU_PredictedCleaned.csv")
df.head()

output_df = pd.DataFrame(columns=df.columns)

grouped = df.groupby(df['Street Address'].str.split().str[0]+" "+df['Street Address'].str.split().str[1])

for _, group in grouped:
    if len(group) > 1:
        sum_guest_count = group['Sum_Guest Count'].sum()
        daily_waste_ton = sum_guest_count / 8626.465887
        daily_waste_per_person = 0.5 / daily_waste_ton
        new_row = group.iloc[0].copy()
        new_row['Sum_Guest Count'] = sum_guest_count
        new_row['Daily_Waste_Ton'] = daily_waste_ton
        output_df = output_df.append(new_row, ignore_index=True)
    else:
        output_df = output_df.append(group, ignore_index=True)

output_df
output_df.to_csv('output.csv', index=False)
```
